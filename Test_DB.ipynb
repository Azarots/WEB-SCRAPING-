```python
import pandas as pd
from matplotlib import pyplot as plt
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///Countries_DB.db')
```

```python
data = pd.read_csv('countries.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.year
```

```python
def population_stats(df, country):
    country_df = df[df["country"] == country]
    min_population = country_df["population"].min()
    max_population = country_df["population"].max()
    avg_population = int(country_df["population"].mean())
    return min_population, max_population, avg_population
```

```python
country = "Cuba"
```

```python
min_population, max_population, avg_population = population_stats(df, country)
print("Min population of", country, ":", min_population)
print("Max population of", country, ":", max_population)
print("Avg population of", country, ":", avg_population)
```
